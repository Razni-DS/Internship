```python
#Q10- Write a python program to scrape the salary data for Data Scientist designation.

url = "https://www.ambitionbox.com/"
driver.get(url)
```

```python
salaries_button=driver.find_element_by_xpath('/html/body/div[1]/nav/nav/a[4]')
salaries_button.click()
```

```python
search_button=driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input')
search_button.send_keys("Data Scientist")
```

```python
company=driver.find_elements_by_xpath('//div[@class="name"]/a')
```

```python
No_of_Salaries=driver.find_elements_by_xpath('//div[@class="name"]/span')
```

```python
average_salary=driver.find_elements_by_xpath('//p[@class="averageCtc"]')
```

```python
min_salary=driver.find_elements_by_xpath('//div[@class="value body-medium"]')
```

```python
max_salary=driver.find_elements_by_xpath('//div[@class="value body-medium"]')
```

```python
experience=driver.find_elements_by_xpath('//div[@class="salaries sbold-list-header"]')
```

```python
Comp=[]
Salary=[]
Avg=[]
mins=[]
maxs=[]
exp=[]

for i in company:
    Comp.append(i.text)
    
for i in No_of_Salaries:
    Salary.append(i.text)
    
for i in average_salary:
    Avg.append(i.text)
    
for i in min_salary:
    mins.append(i.text)
    
for i in max_salary:
    maxs.append(i.text)
    
for i in experience:
    exp.append(i.text.split()[3:])
```

```python
df=pd.DataFrame({'Company Name':Comp[:10],'No. of Salaries':Salary[:10],'Average Salary':Avg[:10],'Min Salary':mins[:10],'Max Salary':maxs[:10],"Experience rqd":exp[:10]})
df
```
