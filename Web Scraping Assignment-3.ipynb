### Q2- In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
#first we will collect URL's for 3 pages and then go to each URL to scrap details

product_url=[]
for i in range(0,2):
    url_page=driver.find_elements_by_xpath("//a[@class='a-link-normal s-no-outline']")
    for i in url_page:
        product_url.append(i.get_attribute('href'))
        
    #clicking the next button
    next_b=driver.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']") 
    url=next_b.get_attribute('href')
    driver.get(url)
    time.sleep(3)
```

```python
#details to be scraped:

Brand_name=[]
Product_name=[]
Price=[]
Return_Exchange=[]
Expected_Delivery=[]
Availability=[]
```

```python
for i in product_url:
    driver.get(i)
    time.sleep(2)
    
    #scraping Brand name
    try:
        brand=driver.find_element_by_xpath("//div[@id='bylineInfo_feature_div']/div/a")
        Brand_name.append(brand.text)
    except NoSuchElementException as e:
        Brand_name.append("-")
        
    #scraping Product name:
    try:
        prod=driver.find_element_by_xpath("//span[@id='productTitle']")
        Product_name.append(prod.text)
    except NoSuchElementException as e:
        Product_name.append("-")
        
    #Scraping the Price:
    try:
        pri=driver.find_element_by_xpath("//span[@class='a-price a-text-price a-size-medium apexPriceToPay']")
        Price.append(pri.text)
    except NoSuchElementException as e:
        Price.append("-")
        
    #Scraping Return/Exchange:
    try:
        ret=driver.find_element_by_xpath("//div[@data-name='RETURNS_POLICY']/span/div[2]/a")
        Return_Exchange.append(ret.text)
    except NoSuchElementException as e:
        Return_Exchange.append("-")
        
    #Scraping Expected Delivery:
    try:
        deli=driver.find_element_by_xpath("//div[@id='mir-layout-DELIVERY_BLOCK-slot-DELIVERY_MESSAGE']/b")
        Expected_Delivery.append(deli.text)
    except NoSuchElementException as e:
        Expected_Delivery.append("-")
        
    #scraping the Availability:
    try:
        avail=driver.find_element_by_xpath("//div[@id='availability']/span")
        Availability.append(avail.text)
    except NoSuchElementException as e:
        Availability.append("-")
```

```python
#making the dataframe
TV=pd.DataFrame({})
TV['Name'] = Product_name
TV['Brand'] = Brand_name
TV['Price'] = Price
TV['Return'] = Return_Exchange
TV['Expected_Delivery'] = Expected_Delivery
TV['Availability'] = Availability
TV['Urls'] = product_url

TV
```

```python
#Saving to csv
TV.to_csv('Amazon_TV.csv'.format(item))
```

```python
driver.close()
```

### Q4-Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
driver=webdriver.Chrome("/Users/sraazi_reni/Downloads/chromedriver")
#open a website
url = "https://www.flipkart.com/"
driver.get(url)

smartphone= input(" Enter the name of Smartphone : ")
```

```python
#to close the login pop up

close_login = driver.find_element_by_xpath("//div[@class='_2QfC02']//button").click()
```

```python
search = driver.find_element_by_xpath("//div[@class='_3OO5Xc']//input")
search.send_keys(smartphone)

btn = driver.find_element_by_xpath("/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button")
btn.click()
time.sleep(5)
```

```python
product_url=[]

url_page=driver.find_elements_by_xpath("//a[@class='_1fQZEK']")
for i in url_page:
    product_url.append(i.get_attribute('href'))
len(product_url)
```

```python
Brand=[]
Name=[]
Colour=[]
RAM=[]
Storage_ROM=[]
Back_Cam=[]
Front_Cam=[]
Display_size=[]
Battery_Cap=[]
Price=[]
```

```python
for i in product_url:
    driver.get(i)
    time.sleep(2)
    
    #read more
    try:
        read_more = driver.find_element_by_xpath('//button[@class="_2KpZ6l _1FH0tX"]')     
        read_more.click()
    except NoSuchElementException:
        print("Exception occured while moving to next page")
    
    #scraping Brand name
    try:
        brand=driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        Brand.append(brand.text.split()[0])
    except NoSuchElementException as e:
        Brand.append("-")
        
    #scraping Product name:
    try:
        prod=driver.find_element_by_xpath('//div[@class="_3k-BhJ"][1]/table/tbody/tr[3]/td[2]/ul/li')     
        Name.append(prod.text)
    except NoSuchElementException as e:
        Name.append("-")
        
    #Scraping color:
    try:
        clr=driver.find_element_by_xpath('//div[@class="_3k-BhJ"][1]/table/tbody/tr[4]/td[2]/ul/li')      
        Colour.append(clr.text)
    except NoSuchElementException as e:
        Colour.append("-")
        
     #Scraping RAM data
    try:
        ram = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][4]/table[1]/tbody/tr[2]/td[2]/ul/li')                
        RAM.append(ram.text)
    except NoSuchElementException:
        RAM.append('-')
        
    #scraping the storage:
    try:
        rom=driver.find_element_by_xpath('//div[@class="_3k-BhJ"][4]/table[1]/tbody/tr[1]/td[2]/ul/li')        
        Storage_ROM.append(rom.text)
    except NoSuchElementException:
        Storage_ROM.append("-")
                                         
    #Scraping Primary camera data
    try:                                                                                    
        b_cam = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][5]/table[1]/tbody/tr[2]/td[2]/ul/li')
        Back_Cam.append(b_cam.text)
    except NoSuchElementException:
        Back_Cam.append('-')
        
        
    #Scraping secondary camera data
    try:                                                                                    
        f_cam = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][5]/table[1]/tbody/tr[6]/td[1]')
        if f_cam != "Secondary Camera" : 
            if driver.find_element_by_xpath('//div[@class="_3k-BhJ"][5]/table[1]/tbody/tr[5]/td[1]').text == "Secondary Camera":
                f_cam = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][5]/table[1]/tbody/tr[5]/td[2]/ul/li')
            else :
                raise NoSuchElementException
        else :
            f_cam = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][5]/table[1]/tbody/tr[6]/td[2]/ul/li')
        Front_Cam.append(f_cam.text)
    except NoSuchElementException as e:
        Front_Cam.append('-')
                                         
            
      #Scraping Display size data 
    try:
        disp = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][2]/div')
        if disp.text != "Display Features" : raise NoSuchElementException
        disp = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][2]/table[1]/tbody/tr[1]/td[2]/ul/li')  
        Display_size.append(disp.text)
    except NoSuchElementException as e:
        Display_size.append('-')
                                    
    #Scraping battery capacity data
    try:
        if driver.find_element_by_xpath('//div[@class="_3k-BhJ"][10]/div').text != "Battery & Power Features" :
            if driver.find_element_by_xpath('//div[@class="_3k-BhJ"][9]/div').text == "Battery & Power Features" :
                bat_cap = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][9]/table/tbody/tr/td[1]')
                if bat_cap.text != "Battery Capacity" : raise NoSuchElementException
                bat_caps = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][9]/table/tbody/tr/td[2]/ul/li')                
            elif driver.find_element_by_xpath('//div[@class="_3k-BhJ"][8]/div').text == "Battery & Power Features" :
                bat_cap = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][8]/table/tbody/tr/td[1]')
                if bat_cap.text != "Battery Capacity" : raise NoSuchElementException
                bat_caps = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][8]/table/tbody/tr/td[2]/ul/li')
            else:
                raise NoSuchElementException
        else :
            bat_cap = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][10]/table/tbody/tr/td[1]')
            if bat_cap.text != "Battery Capacity" : raise NoSuchElementException
            bat_caps = driver.find_element_by_xpath('//div[@class="_3k-BhJ"][10]/table/tbody/tr/td[2]/ul/li')              
        Battery_Cap.append(bat_caps.text)
    except NoSuchElementException as e:
        Battery_Cap.append('-')
        
        
        
    #Scraping Price data
    try:
        prc = driver.find_element_by_xpath('//div[@class="_30jeq3 _16Jk6d"]')      
        Price.append(prc.text)
    except NoSuchElementException as e:
        Price.append('-')
```

### Q5-  Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
driver=webdriver.Chrome("/Users/sraazi_reni/Downloads/chromedriver")
delay = WebDriverWait(driver, 10)
driver.get("https://www.google.com/maps")
delay.until(EC.element_to_be_clickable((By.ID, "searchboxinput"))).send_keys("mirdif")
delay.until(EC.element_to_be_clickable((By.ID, "searchbox-searchbutton"))).click()
time.sleep(5)

try:
    url = driver.current_url
    print("URL Extracted: ", url)
    latitude_longitude = re.findall(r'@(.*)data',url)
    if len(latitude_longitude):
        lat_lng_list = latitude_longitude[0].split(",")
        if len(lat_lng_list)>=2:
            latitude = lat_lng_list[0]
            longitude = lat_lng_list[1]
        print("Latitude = {}, Longitude = {}".format(latitude, longitude))

except Exception as e:
        print("Error: ", str(e))
```

### Q6- Write a program to scrap details of all the funding deals for second quarter (i.e Jan 21 – March 21) from trak.in.

```python
driver=webdriver.Chrome("/Users/sraazi_reni/Downloads/chromedriver")
driver.get('https://trak.in/')
time.sleep(5)
funding = driver.find_element_by_xpath('//li[@id="menu-item-51510"]/a').get_attribute('href')
driver.get(funding)
```

```python
#Creating empty lists
fund_deals = {}
fund_deals['Date'] = []
fund_deals['Startup Name'] = []
fund_deals['Industry OR Vertical'] = []
fund_deals['Sub-Vertical'] = []
fund_deals['Location'] = []
fund_deals['Investor'] = []
fund_deals['Investment Type'] = []
fund_deals['Amount(in USD)'] = []
```

```python
#Clicking on tablepress
for i in range(54,57):

    # Scraping data of Date
    date_tags = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[2]'.format(i))
    for date in date_tags:
        fund_deals['Date'].append(date.text)

    # Scraping data of Name
    name_tags = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[3]'.format(i))
    for name in name_tags:
        fund_deals['Startup Name'].append(name.text)
    
    # Scraping data of Industry
    ind_tags = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[4]'.format(i))
    for n in ind_tags:
        fund_deals['Industry OR Vertical'].append(n.text)
    
    # Scraping data of Sub-Vertical
    sv_tags = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[5]'.format(i))
    for sv in sv_tags:
        fund_deals['Sub-Vertical'].append(sv.text)

    # Scraping data of Location
    loc_tags = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[6]'.format(i))
    for loc in loc_tags:
        fund_deals['Location'].append(loc.text)
    
    # Scraping data of Investor
    inv_tags = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[7]'.format(i))
    for inv in inv_tags:
        fund_deals['Investor'].append(inv.text)
        
    # Scraping data of Investment Type
    invt_tags = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[8]'.format(i))
    for invt in invt_tags:
        fund_deals['Investment Type'].append(invt.text)
    
    # Scraping data of Amount
    amt_tags = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[9]'.format(i))
    for amt in amt_tags:
        fund_deals['Amount(in USD)'].append(amt.text)
```

### Q7-Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
Product_Name=[]
Price=[]
Processor=[]
Memory=[]
Dimensions=[]
Graphical=[]
```

```python
# Scraping the name of laptops
name_tags = driver.find_elements_by_xpath("//table[@id='summtable']//tr//td[1]")
for name in name_tags:
    Product_Name.append(name.text)
 
    
#Scraping price
try:
    price = driver.find_elements_by_xpath("//td[@class='smprice']")
    for p in price:
        Price.append(p.text.replace('₹','Rs '))
except NoSuchElementException:
    Price.append('-')

#Scraping Processor
try:
    proc = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[5]//td[3]")
    for p in proc:
        Processor.append(p.text)
except NoSuchElementException:
    Processor.append('-')


#Scraping data of memory
try:
    memo = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[6]//td[3]")
    for m in memo:
        Memory.append(m.text)
except NoSuchElementException:
    Memory.append('-')


#Scraping data of dimensions
try:
    dim = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[8]//td[3]")
    for d in dim:
        Dimensions.append(d.text)
except NoSuchElementException:
    Dimensions.append('-')


#Scraping data of Graph processor
try:
    graph = driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[9]//td[3]")
    for g in graph:
        Graphical.append(g.text)
except NoSuchElementException:
    Graphical.append('-')
```

```python
print(len(Product_Name),len(Price),len(Processor),len(Memory),len(Dimensions),len(Graphical))
```

```python
Gaming=pd.DataFrame({'Product Name':Product_Name,'Price':Price,"Processor":Processor,"Memory":Memory,"Dimensions":Dimensions,"Graphical Processor":Graphical})
Gaming
```

```python
driver=webdriver.Chrome("/Users/sraazi_reni/Downloads/chromedriver")
driver.get('https://www.forbes.com/?sh=69e6b8c92254')
time.sleep(5)
#Click explore
button = driver.find_element_by_xpath("//button[@class='icon--hamburger']")
button.click()
time.sleep(2)

#click billionare
billionare = driver.find_element_by_xpath("/html/body/div[1]/header/nav/div[3]/ul/li[1]")
billionare.click()
time.sleep(2)

#click world billionare
world_bill= driver.find_element_by_xpath("/html/body/div[1]/header/nav/div[3]/ul/li[1]/div[2]/ul/li[2]/a")
world_bill.click()
time.sleep(2)
```

```python
while(True):
    #Scraping the data of rank
    rank= driver.find_elements_by_xpath("//div[@class='rank']")
    for r in rank:
        Rank.append(r.text)
    time.sleep(1)
    
    #Scraping the names
    name= driver.find_elements_by_xpath("//div[@class='personName']//div")
    for n in name:
        Name.append(n.text)
    time.sleep(1)
    
    #scraping net_worth of billionaires 
    net= driver.find_elements_by_xpath("//div[@class='netWorth']//div[1]")
    for n in net:
        N_worth.append(n.text)
    time.sleep(1)
    
    #Scraping age
    age= driver.find_elements_by_xpath("//div[@class='age']//div")
    for a in age:
        Age.append(a.text)
    time.sleep(1)
    
    #Scraping citizenship
    citizen= driver.find_elements_by_xpath("//div[@class='countryOfCitizenship']")
    for c in citizen:
        Citizenship.append(c.text)
    time.sleep(1)
    
    #Scraping source of income
    source= driver.find_elements_by_xpath("//div[@class='source']")
    for s in source:
        Source.append(s.text)    
    time.sleep(1)
    
    
    #Scraping Industry
    industry= driver.find_elements_by_xpath("//div[@class='category']//div")
    for i in industry:
        Industry.append(i.text)
    time.sleep(1)
    
    #Clicking on next button
    try:
        next_button = driver.find_element_by_xpath("//button[@class='pagination-btn pagination-btn--next ']")
        next_button.click()
    except:
        break
        
#Scraping net worth
Net_Worth = []
for i in range(0,len(N_worth),2):
    Net_Worth.append(N_worth[i])
```
